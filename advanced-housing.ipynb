```python
import pandas as pd
import pandas_profiling 
melbourne_test = pd.read_csv("../input/house-prices-advanced-regression-techniques/test.csv")
melbourne_data = pd.read_csv("../input/house-prices-advanced-regression-techniques/train.csv")
```

```python
y=melbourne_data.SalePrice
from sklearn.ensemble import RandomForestClassifier
y.head()
```

```python
melbourne_features = ['LotArea', 'LotFrontage', 'HouseStyle', 'BldgType', 'OverallQual', 'OverallCond','GrLivArea','GarageArea','YearBuilt','TotalBsmtSF']
```

```python
melbourne_data.head()
```

```python
X = melbourne_data[melbourne_features]
X_test= melbourne_test[melbourne_features]
X.head()
```

```python

features = [ "HouseStyle", "BldgType"]
X = pd.get_dummies(X)
X_test = pd.get_dummies(X_test)
```

```python
Fin=[]
for val in range(0,len(X_test)):
    Fin.append(0)
print(len(Fin))
print(len(X_test))
X_test['HouseStyle_2.5Fin']=Fin
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
my_imputer = SimpleImputer()
X = my_imputer.fit_transform(X)
X_test = my_imputer.fit_transform(X_test)

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = RandomForestRegressor(random_state=1,n_estimators=1000)

# Fit model
melbourne_model.fit(X, y)
```

```python
predictions = melbourne_model.predict(X_test)
```

```python
output = pd.DataFrame({'Id': melbourne_test.Id, 'SalePrice': predictions})
output.to_csv('my_submission.csv', index=False)
print("Done")
```
